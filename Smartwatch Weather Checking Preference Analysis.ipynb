```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('weather-check.csv')

# Data Cleaning and Preprocessing
# Drop rows with missing values
df.dropna(inplace=True)

# Convert categorical variables to appropriate data types
df['What is your gender?'] = df['What is your gender?'].astype('category')
df['US Region'] = df['US Region'].astype('category')

# Data Analysis
# Focus on the responses to the question asking about checking weather on a smartwatch
smartwatch_usage = df["If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?"].value_counts(normalize=True)

# Visualize the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=smartwatch_usage.index, y=smartwatch_usage.values)
plt.title("Likelihood of Checking Weather on a Smartwatch")
plt.xlabel("Likelihood")
plt.ylabel("Percentage")
plt.xticks(rotation=45, ha='right')  # Rotate labels and align to the right
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

# Explore whether preferences vary based on demographic factors
# For example, you can analyze smartwatch usage by age, gender, income, and region using groupby and aggregation functions
# Here's an example of analyzing usage by gender
usage_by_gender = df.groupby('What is your gender?')["If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?"].value_counts(normalize=True)

# Visualize the distribution of usage by gender
plt.figure(figsize=(12, 6))
sns.barplot(x=usage_by_gender['Male'].index, y=usage_by_gender['Male'].values)
plt.title("Smartwatch Usage among Males")
plt.xlabel("Likelihood")
plt.ylabel("Percentage")
plt.xticks(rotation=45, ha='right')  # Rotate labels and align to the right
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=usage_by_gender['Female'].index, y=usage_by_gender['Female'].values)
plt.title("Smartwatch Usage among Females")
plt.xlabel("Likelihood")
plt.ylabel("Percentage")
plt.xticks(rotation=45, ha='right')  # Rotate labels and align to the right
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
