```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
weather_data = pd.read_csv("weather-check.csv")

# Display the first few rows of the dataset to understand its structure
print(weather_data.head())

#Age Distribution
age_distribution = weather_data['Age'].value_counts().sort_index()
age_distribution.plot(kind='bar', figsize=(10, 6), title='Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

#Gender Distribution
gender_distribution = weather_data['What is your gender?'].value_counts()
gender_distribution.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Gender Distribution')
plt.ylabel('')
plt.show()

#US Region Distribution
region_distribution = weather_data['US Region'].value_counts()
region_distribution.plot(kind='bar', figsize=(10, 6), title='US Region Distribution')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```
